```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(2340, 100, input_length=10))
# The first LSTM layer must return the full sequence to be fed into the next LSTM layer
model.add(LSTM(units=256, return_sequences=True))
# The second LSTM layer will now correctly receive a 3D input
model.add(LSTM(units=124))
model.add(Dense(2340, activation='softmax'))
```

what is the work of embeddied layer
here we have 28 words for each sentence but after passing from embedded layer we will have 100 numbers for each word i.e 1 row of 28 words and each word wiull be represented as 200 different numbers for entensively leasrning

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

we are having 28 words each sentence so when we will input our dataset in the lstm layer it will recur 28 time i.e. we will have a totsal of 28 timestamps for ewach input and each word will go in as 100 numbers therefor we will have 1099 nodes with 100 weightsd for all so total 109900 weights



also from the lstm layer after runnign for 28 timestamps we will get ht as output the ht wil have 150 terms as ht is an array whoose length ois always equal to the number of nodes in the lstm layer



then those 150 wala ht willl go to the softmax layer with 1099 nodes this 150 units will go tothe layer with 1099 nodes and thern finally we will get an array of10989 probabilities corresponding to each word
